```python
def to_model(method):
    def model(x):
        global x_
#         print(x)
        x = np.mean(x, axis=(-3, -2))
#         print(x)
        x_ = x
        x = method(x)
#         print(x_)
        x = (x-np.expand_dims(np.mean(x, axis=-1), -1))/(np.expand_dims(np.std(x, axis=-1), -1)+1e-6)
        return x
    return model
```
